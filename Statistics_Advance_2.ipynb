**Q1)  What is hypothesis testing in statistics?**

Ans- Hypothesis testing is a statistical method used to make decisions or inferences about a population based on a sample. It involves formulating a claim (hypothesis) and then using sample data to determine whether there is enough evidence to reject the claim.

For example, if a company claims that its battery lasts 10 hours on average, hypothesis testing can be used to verify this claim using sample data.

**Q2)  What is the Null Hypothesis, and How Does It Differ from the Alternative Hypothesis?**

Ans- The null hypothesis (H₀) assumes no change, while the alternative hypothesis (H₁) suggests a change.Hypothesis testing aims to determine whether to reject H₀ or fail to reject H₀ based on the sample data.

**Null Hypothesis (H₀):** It represents the status quo or a claim that there is no effect, no difference, or no relationship. It assumes that any observed differences in the data are due to random chance.
Example- H₀: The average battery life is 10 hours.

**Alternative Hypothesis (H₁ or Ha):** It represents what we seek to prove, suggesting that there is an effect, a difference, or a relationship.
Example- Ha: The average battery life is not 10 hours.


**Q3) What is the significance level in hypothesis testing, and why is it important?**

Ans- The **significance level** (denoted as **α**) in hypothesis testing represents the probability of rejecting the null hypothesis (**H₀**) when it is actually true. It is a threshold set by the researcher to determine how much evidence is needed to conclude that an observed effect is statistically significant.

**It is important because :**
1. **Controls Type I Error (False Positive)**  
   * A significance level of **α = 0.05** means there is a 5% chance of incorrectly rejecting the null hypothesis when it is actually true.
   * Lower α values (e.g., 0.01) reduce the likelihood of Type I errors but increase the risk of Type II errors (failing to reject a false H₀).

2. **Determines Statistical Significance**  
   * If the **p-value** (probability of observing the data assuming H₀ is true) is **less than α**, we reject H₀, indicating strong evidence against it.
   * If **p-value ≥ α**, we fail to reject H₀, meaning there is insufficient evidence to support the alternative hypothesis (**H₁**).

3. **Balances Risk and Confidence**  
   * A lower **α** increases confidence in results but requires stronger evidence.
   * A higher **α** allows more flexibility but increases the risk of false positives.

**Common Significance Levels**

 **0.05 (5%)** → Most common in social sciences.

 **0.01 (1%)** → Used in fields requiring high precision (e.g., medicine, physics).

 **0.10 (10%)** → Sometimes used in exploratory studies.

**Q4)  What does a P-value represent in hypothesis testing?**

Ans- The p-value measures the probability of obtaining the observed results (or more extreme results) if the null hypothesis (H₀) is true.It quantifies the strength of evidence against H₀

**Q5)  How do you interpret the P-value in hypothesis testing?**

Ans- The p-value in hypothesis testing represents the probability of obtaining results at least as extreme as the observed data, assuming the null hypothesis (H₀) is true. It quantifies the strength of evidence against H₀.

**Interpretation of the P-value in hypothesis testing:**

**1) Small p-value (≤ α) →** Reject H₀ (Strong evidence against H₀)
* If the p-value is smaller than or equal to the chosen significance level (α, e.g., 0.05), it suggests that the observed data is unlikely under H₀.
* This means there is enough evidence to support the alternative hypothesis (H₁).

**2) Large p-value (> α) →** Fail to reject H₀ (Insufficient evidence)
* If the p-value is greater than α, the data does not provide strong enough evidence against H₀.
* This does not prove that H₀ is true; it only means we do not have enough evidence to reject it.

**Example Interpretation (α = 0.05)**
* p = 0.03 → Since p < 0.05, we reject H₀ and conclude that there is significant evidence for H₁.
* p = 0.08 → Since p > 0.05, we fail to reject H₀, meaning the evidence is insufficient to support H₁.

**Q6) What are Type 1 and Type 2 errors in hypothesis testing?**

Ans- In hypothesis testing, we make decisions based on sample data, but there is always a chance of making an error. The two possible errors are **Type 1 Error** and **Type 2 Error**.

**1. Type 1 Error (False Positive)**
* **Definition:** Rejecting the **null hypothesis (H₀)** when it is actually **true**.
* **Symbol:** Represented by **α** (significance level).
* **Probability:** The risk of committing a Type 1 error is **equal to α**, which is usually set at 0.05 (5%) or lower.

**Example:**
* A court wrongly convicts an **innocent person** (rejecting H₀: "The person is innocent" when it's true).
* A medical test **falsely detects** a disease in a healthy person.

**2. Type 2 Error (False Negative)**
* **Definition:** Failing to reject the **null hypothesis (H₀)** when it is actually **false**.
* **Symbol:** Represented by **β** (the probability of making a Type 2 error).
* **Power of the Test:** The probability of **correctly rejecting** a false H₀ is **1 - β** (called statistical power).

**Example:**
* A guilty person is **wrongly acquitted** in court (failing to reject H₀: "The person is innocent" when it's false).
* A medical test **fails to detect** a disease in a sick person.

**Q7) What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**

Ans- In hypothesis testing, the choice between a one-tailed and a two-tailed test depends on the direction of the effect being tested.

**One-Tailed Test (Directional Test)**
* Tests for an effect in one specific direction (either greater than or less than).
* ₀ (null hypothesis): The population parameter has no effect.
* H₁ (alternative hypothesis): The population parameter is either greater than or less than a specific value.

**Two-Tailed Test (Non-Directional Test)**
* Tests for an effect in both directions (either greater or smaller).
* H₀: No effect.
* H₁: The parameter is different (can be higher or lower).

**Q8) What is the Z-test, and when is it used in hypothesis testing?**

Ans- The Z-test is a statistical test used to determine if the mean of a sample is significantly different from a known population mean.

**Use the Z-test when:**
* Sample size (n) is large (≥ 30).
* Population standard deviation (σ) is known.
* Data is normally distributed or sample size is large enough for CLT to apply.

**Examples of Z-test:**
* One-sample Z-test: Checking if a factory’s machine produces 500g sugar packets on average.
* Two-sample Z-test: Comparing the mean scores of two schools to see if they perform differently.

**Q9)  How do you calculate the Z-score, and what does it represent in hypothesis testing?**

Ans- The Z-score (or standard score) tells how many standard deviations a data point is from the mean.

**Formula:** $$ Z = \frac{X - \mu}{\sigma} $$

Where:
X = Observed value,
μ = Population mean,
σ = Population standard deviation

**Interpretation of Z-score:**
* Z > 0: The value is above the mean.
* Z < 0: The value is below the mean.
* Z = 0: The value is exactly the mean.
* Z = ±1.96: The value is at 95% confidence interval.

**Example Calculation:**
If a student scores 85 on a test with a mean of 75 and a standard deviation of 5:

$$ Z = \frac{85 - 75}{5} = \frac{10}{5} = 2 $$
The student's score is 2 standard deviations above the mean.

**Q10) What is the T-distribution, and when should it be used instead of the normal distribution?**

Ans- The t-distribution is similar to the normal distribution but used when sample size is small (n < 30) or population standard deviation (σ) is unknown.

**Use the t-distribution when:**
* The sample size is small (typically n < 30).
* The population standard deviation (σ) is unknown.
* The data is approximately normally distributed.

**Example of When to Use a T-Test Instead of a Z-Test:**

If we want to test whether the average height of students in a class (n = 15) differs from the national average, but we don't know the population standard deviation, we use a t-test instead of a Z-test.

**Q11) What is the difference between a Z-test and a T-test?**

Ans- Difference Between a Z-test and a T-test:

**1) Z-Test**
* It is used when comparing sample and population means with a known population variance.
* used when sample size is large (n ≥ 30) and population variance (σ²) is known.
* Used Standard Normal Distribution (Z-distribution)
* Always follows a normal distribution.
* Best to compare sample mean to population mean (one-sample Z-test) & two population means (two-sample Z-test).
* Example - Checking if the average height of students in a university (n=100) matches the national average with known variance.

**2) T-test**
* It is used when comparing sample and population means when population variance is unknown.
* used when sample size is small (n < 30) and population variance (σ²) is unknown.
* follows Student's T-Distribution.
* It is more spread out (fatter tails), especially for small samples. Approaches normal distribution as n increases.
* Best for Small sample hypothesis testing and used for one-sample, paired-sample, and independent-sample t-tests.
* Example - Checking if a new drug improves patient recovery time when testing a sample of 15 patients.

**Q12)  What is the T-test, and how is it used in hypothesis testing?**

Ans- A T-test is a hypothesis test used to compare the mean of a sample to a known mean or to compare two sample means when the population standard deviation is unknown.

**Types of T-Tests:**
* **One-Sample T-Test:** Compares the mean of a single sample to a known population mean.

  Example: Checking if the average IQ of a small group differs from 100.
* **Two-Sample T-Test (Independent T-Test):** Compares the means of two independent groups.

  Example: Comparing the average test scores of two different schools.
* **Paired T-Test (Dependent T-Test):** Compares two related samples, like before-and-after measurements.

  Example: Measuring weight loss before and after a fitness program.

**Steps in Hypothesis Testing with T-Test**  
1. **Set Hypotheses:**  
   * H₀ (Null): No difference in means.  
   * H₁ (Alternative): Significant difference in means.  
2. **Choose α (e.g., 0.05).**  
3. **Calculate T-Statistic:**  
   $$ T = \frac{X̄ - μ}{s / \sqrt{n}} $$
4. **Compare T-value with critical value or p-value.**  
5. **Decision:**  
    * **p ≤ α** → Reject H₀ (Significant difference).  
    * **p > α** → Fail to reject H₀ (No significant difference).

**Q13) What is the relationship between Z-test and T-test in hypothesis testing?**

Ans- The **Z-test** and **T-test** both compare sample means in hypothesis testing.  

* **Z-test** is used for **large samples (n ≥ 30)** when the **population standard deviation (σ) is known** and follows a **normal distribution**.  
* **T-test** is used for **small samples (n < 30)** when **σ is unknown**, using the **T-distribution**, which accounts for more variability.  
* As sample size increases, the T-distribution approaches the normal (Z) distribution.
* T-test is more commonly used in real-world scenarios since σ is often unknown.  

**Key Rule:** Use **Z-test for large samples with known σ** and **T-test for small samples or unknown σ**.

**Q14) What is a confidence interval, and how is it used to interpret statistical results?**

Ans- **Confidence Interval (CI) and Its Interpretation**  

A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (e.g., mean or proportion) with a certain level of confidence.  

**Formula for Confidence Interval:**  
For a population mean:  
$$ CI = \bar{x} \pm Z \times \frac{\sigma}{\sqrt{n}}$$

**Interpretation of a 95% Confidence Interval:**  
- A **95% confidence interval** means that if we took many random samples, **95% of the calculated intervals would contain the true population mean**.  
- **Narrower CIs** indicate more precise estimates, while **wider CIs** suggest greater uncertainty.  

**Example Interpretation:**  
If a **95% CI for average height** is **(168 cm, 172 cm)**, we are **95% confident** that the **true population mean height** lies within this range.  

**Q15) What is the margin of error, and how does it affect the confidence interval?**

Ans- The margin of error (ME) represents the uncertainty in a sample estimate. It defines the range within which the true population parameter is likely to lie, based on a given confidence level.

**Formula for Margin of Error:**
$$ME = Z \times \frac{\sigma}{\sqrt{n}}$$

**Impact on Confidence Interval**:
* **Larger ME** → **Wider confidence interval** (more uncertainty).
* **Smaller ME** → **Narrower confidence interval** (more precision).

**Factors Influencing ME**:
* **Sample Size**: Larger sample size reduces ME.
* **Confidence Level**: Higher confidence level increases ME.
* **Standard Deviation**: Greater variability increases ME.

In short, the margin of error controls the width of the confidence interval, with a larger ME implying less precision in the estimate.

**Q16)How is Bayes' Theorem used in statistics, and what is its significance?**

Ans- Bayes' Theorem is a fundamental concept in probability theory that allows you to update the probability of a hypothesis given new evidence. It is used to calculate the posterior probability of an event based on prior knowledge and new data.

**Formula:**
 $$ P(A|B) = \frac{P(B|A) \times P(A)}{P(B)} $$

Where:
* P(A|B) is the posterior probability: Probability of event A given event B.
* P(B|A) is the likelihood: Probability of event B given event A.
* P(A) is the prior probability: Initial probability of event A.
* P(B) is the marginal likelihood: Probability of event B.

**Significance in Statistics:**
* Predictive Modeling: Used to make predictions in classification problems (e.g., in spam filters or medical diagnosis).
* Machine Learning: Forms the basis for Naive Bayes classifiers.
* Decision Making: Helps in refining hypotheses as new evidence becomes available.

**Q17)  What is the Chi-square distribution, and when is it used?**

Ans- The Chi-square (X²) distribution is a probability distribution used for hypothesis testing, especially in tests of goodness of fit, independence, and homogeneity. It is commonly used when the data are categorical and the sample size is large.
* **Degrees of Freedom (df):** The shape of the distribution depends on the number of categories minus one or the number of parameters estimated.

It is used when:
* Goodness of Fit: Tests if a sample matches the population distribution.
* Test of Independence: Checks if two categorical variables are independent.
* Homogeneity of Proportions: Compares proportions across different groups.

**Q18) What is the Chi-square goodness of fit test, and how is it applied?**

Ans- The Chi-square goodness of fit test is used to determine whether a sample data matches an expected distribution. It compares observed frequencies in categories to expected frequencies.

**Formula:**

$$ \chi^2 = \sum \frac{(O_i - E_i)^2}{E_i} $$

Where:
- **O_i** = Observed frequency for category i.
- **E_i** = Expected frequency for category i.
- **Σ** denotes the summation over all categories.

**Flow of Application**:
* Null Hypothesis (H₀): Assumes that the observed data fits the expected distribution.
* Alternative Hypothesis (H₁): Assumes that the observed data does not fit the expected distribution.
* Decision: If the calculated X² value is greater than the critical value from the χ² distribution table (based on degrees of freedom and significance level), reject the null hypothesis.

**Q19)  What is the F-distribution, and when is it used in hypothesis testing?**

Ans- The F-distribution is a probability distribution that arises when comparing two sample variances in hypothesis testing. It is used in tests that compare the variability between group means (like in ANOVA).The F-distribution is skewed to the right, and its shape depends on the degrees of freedom of the numerator (between-group variance) and the denominator (within-group variance).

**Used When:**
* ANOVA (Analysis of Variance): To compare the means of more than two groups.
* Comparing Variances: Used in tests like the F-test to compare variances of two populations.

**Q20)  What is an ANOVA test, and what are its assumptions?**

Ans- ANOVA (Analysis of Variance) is a statistical test used to compare the means of three or more groups to determine if at least one group mean is significantly different from the others.It is of 2 types:
* One-way ANOVA: Tests differences between groups based on one factor.
* Two-way ANOVA: Tests differences based on two factors, with or without interaction between them.

**Assumptions of ANOVA:**
* Independence: The samples are independent of each other.
* Normality: The data in each group are normally distributed.
* Homogeneity of Variance: The variances in the groups are equal (also called homoscedasticity).

**Q21) What are the different types of ANOVA tests?**

Ans- Different types of ANOVA tests:

1) One-Way ANOVA: Compares means of three or more groups based on one factor.
* Example: Comparing the mean scores of students from three different schools.

2) Two-Way ANOVA: Compares means based on two factors, and can also examine the interaction between them.
* Example: Comparing the mean test scores based on study method and gender.

3) Repeated Measures ANOVA: Compares means across multiple time points or conditions for the same participants.
* Example: Measuring the effect of a drug at different time intervals on the same group of patients.

**Q22) What is the F-test, and how does it relate to hypothesis testing?**

Ans- The F-test is used to compare the variances of two populations or to test the overall significance of a model (as in ANOVA).

**Formula :**
$$
F = \frac{\text{Variance of Group 1}}{\text{Variance of Group 2}}
$$

Where the F-statistic follows an F-distribution.

**Application in Hypothesis Testing:**
* Used to compare the variability between two samples (e.g., comparing the variance of heights between two groups).
* Used in ANOVA to test if the group means are significantly different by comparing the variances within and between the groups.
* Null Hypothesis (H₀): Assumes that the variances of the two groups are equal.
* Alternative Hypothesis (H₁): Assumes that the variances of the two groups are different.

```python
#1 Write a Python program to perform a Z-test for comparing a sample mean to a
# known population mean and interpret the results.

import numpy as np
import scipy.stats as stats

# Simulated sample data
np.random.seed(0)
# Sample of 50 data points, mean=100, std=15
sample_data = np.random.normal(100, 15, 50)

population_mean = 100  # Population mean
population_std = 15  # Population standard deviation

# Sample mean and size
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-score formula
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# P-value calculation
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

print(f"Z-score: {z_score:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation of the result
if p_value < 0.05:
    print("Reject the null hypothesis: Sample mean is significantly different from population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
#3 Implement a one-sample Z-test using Python to compare the sample mean with the population mean.

import numpy as np
import scipy.stats as stats

# Sample data (heights of 10 students in cm)
sample_data = [158, 162, 164, 157, 165, 160, 168, 155, 161, 159]

# Known population parameters
population_mean = 160  # Population mean (in cm)
population_std = 10    # Population standard deviation (in cm)

# Calculate the sample mean
sample_mean = np.mean(sample_data)

# Calculate sample size
sample_size = len(sample_data)

# calculation of z-score
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

# Calculate the p-value (for two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))

# Print results
print(f"Sample Mean: {sample_mean}")
print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")

# Conclusion based on significance level (alpha = 0.05)
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#4  Perform a two-tailed Z-test using Python and visualize the decision region on a plot.

import matplotlib.pyplot as plt

# Set the confidence level and critical Z-value
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha / 2)

# Create a plot to visualize the decision region
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x)

plt.plot(x, y, label="Z-distribution")
plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.5, label="Rejection Region")
plt.axvline(x=z_critical, color="black", linestyle="--")
plt.axvline(x=-z_critical, color="black", linestyle="--")

plt.title('Two-Tailed Z-Test')
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

def plot_errors():
    z_critical = stats.norm.ppf(1 - 0.05 / 2)

    # Plot the normal distribution
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x)

    plt.plot(x, y, label="Normal Distribution")

    # Type 1 Error (alpha) - Rejecting the null hypothesis when it's true
    plt.fill_between(x, y, where=(x >= z_critical), color="red", alpha=0.5, label="Type 1 Error (alpha)")

    # Type 2 Error (beta) - Failing to reject the null hypothesis when it's false
    plt.fill_between(x, y, where=(x <= -z_critical), color="blue", alpha=0.5, label="Type 2 Error (beta)")

    plt.axvline(x=z_critical, color="black", linestyle="--")
    plt.axvline(x=-z_critical, color="black", linestyle="--")

    plt.title('Type 1 and Type 2 Errors')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

plot_errors()
```

```python
#6 Write a Python program to perform an independent T-test and interpret the results.

from scipy import stats

# Simulate two independent samples
group1 = np.random.normal(100, 15, 50)
group2 = np.random.normal(105, 15, 50)

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2)

print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("Reject the null hypothesis: The two groups are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference.")
```

```python
#9  Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

def confidence_interval(sample_data, confidence_level=0.95):
    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    sample_std = np.std(sample_data, ddof=1)

    z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
    margin_of_error = z_critical * (sample_std / np.sqrt(sample_size))

    return sample_mean - margin_of_error, sample_mean + margin_of_error

# Calculate the confidence interval
ci_lower, ci_upper = confidence_interval(sample_data)

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.

def calculate_expected_frequencies(observed):
    # Total number of observations (sum of observed frequencies)
    total_observations = sum(observed)

    # Number of categories (length of observed list)
    num_categories = len(observed)

    # Expected frequency for each category (assuming uniform distribution)
    expected_frequency = total_observations / num_categories

    # Create a list of expected frequencies for each category
    expected = [expected_frequency] * num_categories

    return expected

# Example: Observed frequencies for 6 categories (e.g., dice rolls)
observed_frequencies = [12, 9, 10, 8, 13, 8]

# Calculate expected frequencies
expected_frequencies = calculate_expected_frequencies(observed_frequencies)

# Display the result
print("Observed Frequencies:", observed_frequencies)
print("Expected Frequencies:", expected_frequencies)
```

```python
#14 Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution.

import numpy as np
from scipy.stats import chisquare

# Observed frequencies (results of 60 dice rolls)
observed = [12, 9, 10, 8, 13, 8]

# Expected frequencies (since it's a fair die, each face should appear 10 times)
expected = [10, 10, 10, 10, 10, 10]

# Perform the Chi-Square Goodness of Fit Test
chi2_stat, p_value = chisquare(observed, expected)

# Display the result
print("Chi-Square Statistic:", chi2_stat)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The observed data does not fit the expected distribution.")
else:
    print("Fail to reject the null hypothesis: The observed data fits the expected distribution.")
```

```python
#16  Implement an F-test using Python to compare the variances of two random samples.

import numpy as np
from scipy.stats import f

# Function to perform F-test
def f_test(sample1, sample2):
    # Calculate the variances of both samples
    var1 = np.var(sample1, ddof=1)  # Sample variance for sample1
    var2 = np.var(sample2, ddof=1)  # Sample variance for sample2

    # Compute the F-statistic (variance1 / variance2)
    f_statistic = var1 / var2 if var1 > var2 else var2 / var1

    # Degrees of freedom for both samples
    df1 = len(sample1) - 1  # Degrees of freedom for sample1
    df2 = len(sample2) - 1  # Degrees of freedom for sample2

    # Compute the p-value for the F-distribution
    p_value = 1 - f.cdf(f_statistic, df1, df2)

    return f_statistic, p_value

# Example samples
sample1 = [12, 15, 13, 14, 16]
sample2 = [10, 11, 9, 10, 8]

# Perform F-test
f_statistic, p_value = f_test(sample1, sample2)

# Display the result
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: The variances are not significantly different.")
```

```python
#17 Write a Python program to perform an ANOVA test to compare
#means between multiple groups and interpret the results.

import numpy as np
from scipy import stats

# Sample data: scores of students in three different groups
group1 = [50, 55, 60, 65, 70]
group2 = [45, 50, 52, 55, 60]
group3 = [60, 65, 70, 75, 80]

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Display the result
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python
#18 Perform a one-way ANOVA test using Python to compare the means of different groups and plot the results.

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Example Data: Scores of students in three different groups
group1 = [50, 55, 60, 65, 70]
group2 = [45, 50, 52, 55, 60]
group3 = [60, 65, 70, 75, 80]

# Perform one-way ANOVA test
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Display the results
print(f"F-statistic: {f_statistic}")
print(f"P-value: {p_value}")

# Visualization: Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(data=[group1, group2, group3], notch=True)
plt.title('One-Way ANOVA: Group Comparison')
plt.xlabel('Groups')
plt.ylabel('Scores')
plt.xticks([0, 1, 2], ['Group 1', 'Group 2', 'Group 3'])
plt.show()

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference between the group means.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference between the group means.")
```

```python
#19 Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA.

from scipy.stats import shapiro, levene

def check_anova_assumptions(groups):
    # Check Normality using Shapiro-Wilk test
    for i, group in enumerate(groups):
        stat, p_value = shapiro(group)
        print(f"Normality Test for Group {i+1}: p-value = {p_value}")
        if p_value < 0.05:
            print(f"Group {i+1} does not follow normal distribution.")
        else:
            print(f"Group {i+1} follows normal distribution.")

    # Check Equal Variance using Levene's test
    stat, p_value = levene(*groups)
    print(f"Equality of Variance Test (Levene's test): p-value = {p_value}")
    if p_value < 0.05:
        print("The variances are not equal.")
    else:
        print("The variances are equal.")

check_anova_assumptions([group1, group2, group3])
```

```python
#20 Perform a two-way ANOVA test using Python to study the interaction between two factors and visualize the results.

import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
import seaborn as sns
import matplotlib.pyplot as plt

# Simulate Data
np.random.seed(0)

# Factor 1: Food (Fruit, Vegetable)
# Factor 2: Exercise (Low, High)
food = ['Fruit', 'Fruit', 'Vegetable', 'Vegetable']
exercise = ['Low', 'High', 'Low', 'High']

# Simulated calories burned data for each group
fruit_low = np.random.normal(120, 10, 10)   # Mean=120, SD=10, n=10
fruit_high = np.random.normal(150, 10, 10)  # Mean=150, SD=10, n=10
veg_low = np.random.normal(110, 10, 10)     # Mean=110, SD=10, n=10
veg_high = np.random.normal(145, 10, 10)    # Mean=145, SD=10, n=10

# Combine all data into a DataFrame
data = pd.DataFrame({
    'Food': ['Fruit']*10 + ['Fruit']*10 + ['Vegetable']*10 + ['Vegetable']*10,
    'Exercise': ['Low']*10 + ['High']*10 + ['Low']*10 + ['High']*10,
    'Calories_Burned': np.concatenate([fruit_low, fruit_high, veg_low, veg_high])
})

# Step 1: Perform Two-Way ANOVA
model = ols('Calories_Burned ~ C(Food) + C(Exercise) + C(Food):C(Exercise)', data=data).fit()
anova_table = sm.stats.anova_lm(model, typ=2)

# Display ANOVA results
print(anova_table)

# Step 2: Visualization
plt.figure(figsize=(8,6))
sns.boxplot(x='Food', y='Calories_Burned', hue='Exercise', data=data)
plt.title("Two-Way ANOVA: Food Type and Exercise Level on Calories Burned")
plt.show()
```

```python
#24 Perform a hypothesis test for population variance using a Chi-square distribution and interpret the results.

data = np.random.normal(50, 10, 30)  # Mean = 50, Std = 10, N = 30

population_variance = 100

# Perform the chi-square test for population variance
n = len(data)
sample_variance = np.var(data, ddof=1)
chi_square_stat = (n - 1) * sample_variance / population_variance
df = n - 1

# Calculate p-value for two-tailed test
p_value = 2 * min(stats.chi2.cdf(chi_square_stat, df), 1 - stats.chi2.cdf(chi_square_stat, df))

print(f"Chi-square statistic: {chi_square_stat}")
print(f"P-value: {p_value}")
```
