**1. What is the difference between a function and a method in Python?**

Ans- Difference between a function and a method:

Function: A block of reusable code defined using the def keyword, which performs a specific task. Functions are independent of any object.

Method: A function that is associated with an object and is defined inside a class. It requires an instance of the class to be called.

**Example:**

Function: print("Hello!") works anywhere.

Method: "Hello!".lower() works only because it's tied to a string.

**Real-Life Example:**
A function is like a recipe card that anyone can use.
A method is a recipe inside a cookbook that is specific to that book (the object).

**Example in Python:**

```python
# Function
def greet(name):  # Independent function
    return f"Hello, {name}!"

print(greet("Ekta"))

# Method
class Person:
    def __init__(self, name):
        self.name = name

    def greet(self):  # Function inside a class (method)
        return f"Hello, {self.name}!"

person = Person("Ekta")
print(person.greet())
```

**5) What are iterators in Python and how do they differ from iterables?**

Ans -  Difference between iterators and iterables:

**Iterable**: An object capable of returning its elements one at a time, such as lists, tuples, or strings.

**Real-Life Example:**

```python
# Iterable
fruits = ["apple", "banana", "cherry"]

# Iterator
fruit_iterator = iter(fruits)

print(next(fruit_iterator))  # Output: apple
print(next(fruit_iterator))  # Output: banana
```

**6. Explain the concept of generators in Python and how they are defined.**

Ans - Generators in Python are functions that produce values one at a time, allowing you to iterate over them without storing all values in memory.

Generators can be created in two main ways:
1.  **Using a Generator Function**: A generator function is defined like a normal function but uses the 'yield; statement to produce values one at a time.
2.  **Using a Generator Expressio**: A generator expression is similar to a list comprehension but uses parentheses () instead of square brackets [].

**Real-Life Example:**
A vending machine gives one snack at a time (generator), unlike storing all snacks at once (list).




```python
# Using Generator Function
def count_up_to(n):
    count = 1
    print("Using Generator Function")
    while count <= n:
        yield count
        count += 1
for num in count_up_to(3): #calling
    print(num)

#Using Generator Expression:
#gen = (x**2 for x in range(5))
print("Using Generator Expression:")
for value in (x**2 for x in range(5)):
    print(value)
```

**7. What are the advantages of using generators over regular functions?**

Ans -  Advantages of using generators:
* Generators are memory-efficient because they generate items on demand.
* They are faster for large datasets since they don’t create intermediate collections.

**Example in Python:**

```python
# Regular function
def create_list(n):
    return [i for i in range(n)]

# Generator
def create_numbers(n):
    for i in range(n):
        yield i

# Memory usage
print(create_numbers(5))
# Generator object reference (memory location) because it uses yield, which doesn't immediately generate all values at once.
print(create_list(5))
# Full list because it generates and stores all the values at once.
```

**8. What is a lambda function in Python and when is it typically used?**

Ans - A lambda function in Python is a small, anonymous function defined using the 'lambda' keyword.lambda functions are typically used for short-lived operations where we don't want to create a full function with a name.

**Syntax:**
    lambda arguments: expression
    arguments: The input parameters to the function.
    expression: The code that is executed and returned when the lambda function is called.

Lambda functions are typically used for:

1. **Short, one-off functions**: When you need a small function without defining a full function using `def`.
2. **Passing functions as arguments**: In functions like `map()`, `filter()`, and `sorted()`.
3. **Custom sorting**: Defining sorting criteria for lists or collections.
4. **List comprehensions**: Applying simple transformations on elements.
5. **Event handling or callbacks**: For quick, inline functions in event-driven programming.

**Real Life Example:**
writing quick math on a sticky note for immediate use.

**Example in Python:**

```python
square = lambda x: x ** 2
print(square(5))  # Output: 25
```

**10. What is the difference between map(),reduce(), and filter() functions in Python**

Ans - difference between map(), reduce(), and filter() functions:

1. **map()**: Applies a given function to all items in an iterable (like a list) and returns a new iterable (map object) with the results.

  **Syntax**:
            
    map(function, iterable)
    
  **Example**:
     
     numbers = [1, 2, 3, 4]
     result = map(lambda x: x * 2, numbers)
     print(list(result))  # Output: [2, 4, 6, 8]
     

2. **filter()**: Filters elements from an iterable based on a function that returns 'True' or 'False', and returns a new iterable containing only the elements for which the function returned True'.

   **Syntax**:

    filter(function, iterable)

   **Example**:
     
     numbers = [1, 2, 3, 4, 5]
     result = filter(lambda x: x % 2 == 0, numbers)
     print(list(result))  # Output: [2, 4]

3. **reduce()**: Applies a function cumulatively to the items in an iterable, from left to right, to reduce them to a single value (e.g., summing a list of numbers).

**Syntax**:

    from functools import reduce
    reduce(function, iterable)
     
**Example**:
    
     from functools import reduce
     numbers = [1, 2, 3, 4]
     result = reduce(lambda x, y: x + y, numbers)
     print(result)  # Output: 10

**Real-Life Example:**

* map(): Changing the color of all shirts in a wardrobe.
* reduce(): Adding up all grocery prices for the bill.
* filter(): Picking only ripe fruits from a basket.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
          total += num
    print(f"Total sum of even numbers: {total}")
    return total

sum_of_evens([1, 2, 3, 4, 5,6])
```

```python
#5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current = 0
        self.next = 1
        self.index = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.index >= self.terms:
            raise StopIteration
        value = self.current
        self.current, self.next = self.next, self.current + self.next
        self.index += 1
        return value

fib = FibonacciIterator(10)
print(list(fib))  # Output: First 10 Fibonacci numbers
```

```python
#6.Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield f"Power of 2^{i}: {2 ** i}"

for message in powers_of_2(5):
    print(message)
```

```python
#7.Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_lines(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # strip() is used to remove leading/trailing whitespace

file_path = r'example.txt'   # Replace with the actual file path
for line in read_file_lines(file_path):
    print(line)
```

```python
#10.Create a Python program that uses `filter()` to remove all the vowels from a given string.

def not_vowel_string(input_string):
    vowels = "aeiouAEIOU"

    # Use filter to remove vowels
    result = ''.join(filter(lambda x: x not in vowels, input_string))

    return result

input_string = input("Enter any string:")
output_string = not_vowel_string(input_string)
print(f"Original String: {input_string}")
print(f"String without vowels: {output_string}")
```
